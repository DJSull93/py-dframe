```python
import pandas as pd
import numpy as np
import random
import string
np.random.seed(56)
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def sco():
    return np.random.randint(0,100)
```

```python
df = pd.DataFrame([[sco() for i in range(4)] for i in range(10)],
             columns=["국", "수", "사", "과"],
             index=[id() for i in range(10)])
```
